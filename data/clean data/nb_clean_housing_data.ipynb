```python
import pandas as pd
import re

def clean_column_name(col):
    col = col.strip().lower()
    col = re.sub(r"[^\w\s]", "", col)
    col = re.sub(r"\s+", "_", col)
    return col

def clean_housing_file(input_path, output_path):
    df = pd.read_csv(input_path)

    df.columns = [clean_column_name(col) for col in df.columns]

    if "period" in df.columns:
        df["period"] = pd.to_datetime(df["period"], errors="coerce")

    if "pivotable_date" in df.columns:
        df["pivotable_date"] = pd.to_datetime(df["pivotable_date"], errors="coerce")

    if "uri" in df.columns and df["uri"].nunique() == 1:
        df = df.drop(columns=["uri"])

    df["is_partial_record"] = df.isnull().any(axis=1)

    df.to_csv(output_path, index=False)
    print(f"Cleaned file saved to: {output_path}")
```

```python
import os
import pandas as pd
import re

base_path = "/Volumes/dbw_ukhousing_dev/default/housing_files"

def clean_column_name(col):
    col = col.strip().lower()
    col = re.sub(r"[^\w\s]", "", col)
    col = re.sub(r"\s+", "_", col)
    return col

def clean_housing_file(input_path, output_path):
    df = pd.read_csv(input_path)

    df.columns = [clean_column_name(col) for col in df.columns]

    if "period" in df.columns:
        df["period"] = pd.to_datetime(df["period"], errors="coerce")

    if "pivotable_date" in df.columns:
        df["pivotable_date"] = pd.to_datetime(df["pivotable_date"], errors="coerce")

    if "uri" in df.columns and df["uri"].nunique() == 1:
        df = df.drop(columns=["uri"])

    df["is_partial_record"] = df.isnull().any(axis=1)

    df.to_csv(output_path, index=False)
    print(f"Cleaned file saved to: {output_path}")

for file_name in os.listdir(base_path):
    if file_name.endswith(".csv") and "_cleaned" not in file_name:
        input_file = os.path.join(base_path, file_name)
        output_file = os.path.join(base_path, file_name.replace(".csv", "_cleaned.csv"))
        clean_housing_file(input_file, output_file)
```
